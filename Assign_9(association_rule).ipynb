# Applying Apriori Algorithm


# 1) min-support=10% & confidence=1

```python
#importing mlxtend
import mlxtend
from mlxtend.frequent_patterns import apriori
import warnings
warnings.filterwarnings("ignore")
frequent_itemsets=apriori(book,min_support=0.1,use_colnames=True)
frequent_itemsets
```

 This is all association rules which is having a min_support value of 10%

It is going to start with one item set (book) and then slowly it create two item set and then 3 item set i.e (GeogBks, CookBks, ArtBks) which is having 3 item set having support value of 0.1035 which is slightly > 0.10

# Association rules

```python
#Applying association rules on frequent item sets
import mlxtend
from mlxtend.frequent_patterns import association_rules
rules=association_rules(frequent_itemsets,metric="lift",min_threshold=1)
rules
```

```python
#sorting in a decending order
rules.sort_values("lift",ascending=False)
```

Here, 1st association rule is having a high lift ratio value is 2.320186
i.e a person who reads Cookbooks will also read ItalCook.
And confidence for this rule is 26%

Now, there is another association rule having same lift ratio i.e a person who reads ItalCook will also read Cookbooks.
And confidence for this rule is 100%

Therefore, we select this association rule having high confidence.

Similarly,
Here, association rule is having lift ratio value is 2.274247
i.e a person who reads (ChildBks, ArtBks) will also read (GeogBks)	.
And confidence for this rule is 62%

Now, there is another association rule having same lift ratio i.e a person who reads (GeogBks) will also read (ChildBks, ArtBks).
And confidence for this rule is 36%

Therefore, we select association rule having high confidence.

And so on.


```python
rules[rules.lift>1]
```

```python
#visualizing the rules
import matplotlib.pyplot as plt
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

# Applying Apriori Algorithm
# 1) min-support=15% & confidence=75%

```python
#importing mlxtend
import mlxtend
from mlxtend.frequent_patterns import apriori
import warnings
warnings.filterwarnings("ignore")
frequent_itemsets3=apriori(movies1,min_support=0.15,use_colnames=True)
frequent_itemsets3
```

# Association Rules

```python
import mlxtend
from mlxtend.frequent_patterns import association_rules
rules3=association_rules(frequent_itemsets3,metric="lift",min_threshold=0.75)
rules3
```

```python
#sorting in a decending order
rules3.sort_values("lift",ascending=False)
```

```python
rules3[rules3.lift>1]
```

```python
#visualizing the rules
import matplotlib.pyplot as plt
plt.scatter(rules3['support'],rules3['confidence'])
plt.xlabel('Support')
plt.ylabel('Confidence')
plt.show()
```

- Support: It consider only combinations that occur with higher frequency in the dataset.
- Lift > 1: Indicates a rule that is useful in finding consequent items sets.
- Higher the lift ratio ,better is association rules.
- In dataset lift ratio is having similar or very close then decide the association rule based on confidence. Whichever is having more confidence, select that association rule.
