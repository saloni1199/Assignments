# QUESTION 1

A F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions.


     Minitab File : Cutlets.mtw


```python
# Importing Dataset
import pandas as pd
cutlets=pd.read_csv("Cutlets.csv")
cutlets
```

'''null hypothesis  =  there is no any significant difference in the diameter of the cutlet between two units

alternative hypothesis =  there is any significant difference in the diameter of the cutlet between two units'''

```python
unitA=pd.Series(cutlets.iloc[:,0])
unitA
```

```python
unitB=pd.Series(cutlets.iloc[:,1])
unitB
```

```python
import scipy.stats as stats
p_value=stats.ttest_ind(unitA,unitB)
p_value
```

```python
p_value[1]
```

As our p_value value is less than alpha So we reject the null hypothesis.

So, there is any significant difference in the diameter of the cutlet between two units. 

# QUESTION 2

A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.
 
 
    Minitab File: LabTAT.mtw

null hypothesis =  there is no any difference in average TAT among the different laboratories

alternative hypothesis =  there is any difference in average TAT among the different laboratories

```python
import scipy.stats as stats
from scipy.stats import f_oneway
f_oneway(LabTAT["Laboratory 1"], LabTAT["Laboratory 2"], LabTAT["Laboratory 3"], LabTAT["Laboratory 4"])
```

P_value is greater than alpha we accept null hypothesis

So we can say there is no any difference in average TAT among the different laboratories

# QUESTION 3

Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
# importing dataset
buyerratio=pd.read_csv("BuyerRatio.csv")
buyerratio
```

Null Hypothesis: All proportions are equal

Alternative Hypothesis: Not all Proportions are equal

```python
buyerratio1=buyerratio.iloc[0:,1:]
buyerratio1
```

```python
# Changing the data frame format to array format
data = buyerratio1.to_numpy()
data
```

```python
# To find the Chi square value and p_value
import scipy.stats as stats
from scipy.stats import chi2_contingency
chi2_contingency(data)
```

Chi-square value = 1.5959

p_value = 0.6603

degree of freedom = 3

alpha value = 0.05

As the p_value is greater than alpha value we accept the null hypothesis.
So we can say all propertions are equal.
